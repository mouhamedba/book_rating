```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# reading data
path = "books.csv"
Books = pd.read_csv(path,error_bad_lines=False)
Books.info()
```

```python
def uniques(columns):
    return (columns.uniques())
```

```python
for col in list(Books.columns)[1:]:
    print ("distinct elements:")
    print (uniques(Books[col]))
```
